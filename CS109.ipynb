```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/cs109/2014_data/master/diamonds.csv'
# Let's take a look at some data on diamonds
```

```python
diamonds = pd.read_csv(url)
diamonds.head()
```

```python
diamonds.count()
```

```python
diamonds.describe()
```

```python
diamonds['price'].hist(bins = 30, color = 'red', alpha = 0.8)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel ('Items')
plt.style.use('ggplot')
plt.xlim(1000, 18500)
plt.ylim(0, 7000)
```

```python
diamonds['carat'].hist(bins = 10, color = 'blue', figsize = (6, 4), alpha = 1)
plt.xlabel('weight in carats')
plt.xlim (0.2, 2.7)
plt.ylim(0, 26000)
plt.title('Diamonds in Carats')
```

```python
diamonds['price'].plot(kind = 'kde', color = 'orange', alpha = 1, fontsize = 10)
plt.title('Distribution of Price')
plt.xlim(-700, 20500)
```

# Relationship between Carat and Price

```python
diamonds.plot(x = 'carat', y = 'price', kind = 'scatter', title = 'carat vs price', alpha = 1)
plt.xlim(0, 5.5)
plt.ylim(0, 20000)
```

## Using log values

```python
diamonds.plot(x = 'carat', y = 'price', kind = 'scatter', title = 'carat vs price', logx = True, logy = True, alpha = 1)
### alpha = 1
```

```python
### Still plotting carat vs price relationship, alpha = 0.2
```

```python
diamonds.plot(x = 'carat', y = 'price', kind = 'scatter', title = 'carat vs price', alpha = 0.2)
plt.xlim(0, 5.5)
plt.ylim(0, 20000)
```

```python
diamonds.plot(x = 'carat', y = 'price', kind = 'scatter', title = 'carat vs price', logx = True, logy = True, alpha = 0.2)
### alpha = 0.2
```

### Using 'GroupBy' ('color') to present Carat vs Price relationship slightly more emphatically

```python
diamonds.groupby('color').mean()
```

```python
diamonds.groupby('color').plot(x = 'carat', y = 'price', kind = 'scatter', title = 'carat vs price', alpha = 1)
```

### Now showing it using log values

```python
diamonds.groupby('color').plot(x = 'carat', y = 'price', kind = 'scatter', title = 'carat vs price', alpha = 1, logx = True,
                              logy = True)
```

### Using 'clarity' to group by and demonstrate relationship

```python
diamonds.groupby('clarity').plot(x = 'carat', y = 'price', kind = 'scatter', title = 'carat vs price', alpha = 1)
# Not using 'log' to demonstrate here, slope for each group is very similar.
```
