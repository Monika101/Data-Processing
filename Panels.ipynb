```python
import numpy as np
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
companies = ['MSFT', 'GOOG', 'AAPL', 'YHOO', 'AMZN']
starts = '2016-01-01'
ends = '2016 - 12 - 31'
```

```python
panel = data.DataReader(name = companies, data_source = 'google', start = starts, end = ends)
```
