```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data
import datetime as dt
% matplotlib inline
```

# Microsoft Share Price Improvement HF1 2016 vs HF1 2015 - with graphs.

```python
company = 'MSFT'
start = '2010 - 01 - 01'
end = '2017 - 12 -31'

msft = data.DataReader(name = company, data_source = 'google', start = start, end = end)
```

```python
msft.plot()
```

```python
msft.head()
```

```python
msft.plot(y = ['Low', 'High', 'Open'], logx = True)
```

```python
msft2 = data.DataReader(name = company, data_source = 'google', start = '2016 - 01 - 01', end = end)
msft2.plot(y = ['Low', 'High'], loglog = True)
```

```python
msft2.plot(y = ['Low', 'High'])
```

```python
msft_close = data.DataReader(name = company, data_source = 'google', start = '2016 - 01 - 01', end = '2016 - 06 - 30')
```

```python
msft_close['Close'].plot()
```

```python
msft3 = data.DataReader(name = company, data_source = 'google', start = '2015 - 01 - 01', end = '2015 - 12 - 31')
msft3.plot(y = ['Low', 'High'], logx = True)
```

```python
msft3_close = data.DataReader(name = company, data_source = 'google', start = '2015 - 01 - 01', end = '2015 - 06 - 30')
msft3_close['Close'].plot()
```
