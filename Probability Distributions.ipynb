```python
%matplotlib inline 

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np

# special matplotlib argument for improved plots
from matplotlib import rcParams
```

# Random Variables and Probability Distributions

## Binomial Functions

```python
values = 100
probability = 0.3
random_variables = np.arange(0,101) # plot all values between 0 and 10
y = stats.binom.pmf(random_variables, values, probability)
print y
```

```python
plt.plot(random_variables, y, 'o-')
plt.title('Binomial: n=%i , p=%.2f' % (values,probability),fontsize=15)
plt.xlabel('Number of successes')
plt.ylabel('Probability of successes', fontsize=15)
plt.xlim(10, 50) # to show Distribution in a more friendly format
plt.style.use('default')
plt.show()
```

```python
data = stats.binom.rvs(n = 100, p = 0.3, size = 10000)
print "Mean: %g" % np.mean(data)
print "SD: %g" % np.std(data, ddof=1)

plt.figure()
plt.hist(data, bins = 50, normed = True)
plt.xlabel("number of successes")
plt.ylabel("density")
plt.style.use('default')
plt.show()
```

## Poisson

```python
lambda_ = 38.5 # rate parameter lambda
values = np.arange(0, 101)
y = stats.poisson.pmf(values, lambda_)
print y
```

```python
plt.plot(values, y, 'o-')
plt.title('Poisson: $\lambda$ =%i' % lambda_)
plt.xlabel('Number of accidents')
plt.ylabel('Probability of number of accidents')
plt.style.use('fivethirtyeight')
plt.show()
```

```python
data = stats.poisson.rvs(mu = 38.5, loc = 0, size = 1000)
print "Mean: %g" % np.mean(data)
print "SD: %g" % np.std(data, ddof=1)

plt.figure()
plt.hist(data, bins = 30, normed = False) ##### Normed FALSE ######
plt.xlim(10, 65)
plt.xlabel("Number of accidents")
plt.title("Simulating Poisson Random Variables")
plt.show()
```

```python
data = stats.poisson.rvs(mu = 38.5, loc = 0, size = 1000)
print "Mean: %g" % np.mean(data)
print "SD: %g" % np.std(data, ddof=1)

plt.figure()
plt.hist(data, bins = 30, normed = True) ##### Normed TRUE ######
plt.xlim(10, 65)
plt.xlabel("Number of accidents")
plt.title("Simulating Poisson Random Variables")
plt.show()
```

# Continuous Distribution

## Gaussian (Normal)

```python
mu = 1 # mean
sigma = 2 # standard deviation
x = np.arange(-12,12,0.1)

y = stats.norm.pdf(x, 1, 3)
plt.plot(x, y)
plt.title('Normal: $\mu$=%.1f, $\sigma$=%.1f' % (mu, sigma))
plt.xlabel('x')
plt.ylabel('Probability density')
plt.show()
```

## Exponential

```python
lam = 0.35
x = np.arange(0, 15, 0.1)
y = lam * np.exp(-lam * x) # could also use stats.expon.pdf
plt.plot(x,y)
plt.title('Exponential: $\lambda$ =%.2f' % lam)
plt.xlabel('x')
plt.ylabel('Probability density')
plt.xlim(-1, 14)
plt.show()
```
