```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data
import datetime as dt
% matplotlib inline
```

# Microsoft Share Price Improvement HF1 2016 vs HF1 2015 - with graphs.

```python
company = 'MSFT'
start = '2010 - 01 - 01'
end = '2017 - 12 -31'

msft = data.DataReader(name = company, data_source = 'google', start = start, end = end)
```

```python
msft.plot()
plt.style.use('ggplot')
```

```python
msft.head()
```

```python
#Rough overview from 2010 onwards to show overall share price growth

plt.style.use('fivethirtyeight')
msft.plot(y = ['Low', 'High'])
```

```python
# For 2016 only

msft_2016 = data.DataReader(name = company, data_source = 'google', start = '2016 - 01 - 01', end = '2016 - 12 - 31')
msft_2016.plot(y = ['Low', 'High'])
plt.style.use('ggplot')
```

```python
# For 2015 only

msft_2015 = data.DataReader(name = company, data_source = 'google', start = '2015 - 01 - 01', end = '2015 - 12 - 31')
msft_2015.plot(y = ['Low', 'High'])
plt.style.use('ggplot')
```

```python
#2016 first half

msft_close = data.DataReader(name = company, data_source = 'google', start = '2016 - 01 - 01', end = '2016 - 06 - 30')
```

```python
msft_close['Close'].plot()
plt.style.use('ggplot')
```

```python
#2015 first half

msft_2015_close = data.DataReader(name = company, data_source = 'google', start = '2015 - 01 - 01', end = '2015 - 06 - 30')
msft_2015_close['Close'].plot()
plt.style.use('ggplot')
```
